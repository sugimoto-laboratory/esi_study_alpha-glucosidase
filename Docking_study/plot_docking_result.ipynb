```python
import numpy as np
import pandas as pd
from matplotlib import cm
from matplotlib import pyplot as plt
```

```python
df_a_glu = pd.read_csv('a-glu_docking_result_1217.csv')
vina_affinity_a_glu = df_a_glu['vina_affinity']
pic50_a_glu = df_a_glu['pIC50']
```

```python
# 説明変数と目的変数のデータ点の散布図をプロット
fig = plt.figure(figsize=(5,5))
fig.suptitle('correlation', fontsize=15)

plt.scatter(vina_affinity_a_glu ,pic50_a_glu,  color='blue', marker='.' , label='training set')

plt.xlabel('vina affinity')  # x軸のラベル
plt.ylabel('pIC50')  # y軸のラベル
#plt.xlim(2.5, 7.5)
#plt.ylim(2.5, 7.5)
plt.legend(fontsize=14)
```

```python
df_kampo_ESI = pd.read_csv('ESI_kampo_docking_result_1220.csv')
vina_affinity_kampo_ESI = df_kampo_ESI['vina_affinity']
pic50_kampo_ESI = df_kampo_ESI['pIC50']
p_ESI = len(vina_affinity_kampo_ESI[vina_affinity_kampo_ESI <= -8.0])/ len(vina_affinity_kampo_ESI)
print(f"total: {len(vina_affinity_kampo_ESI)}")
print(f"fit rate of ESI model is {p_ESI}")

df_kampo_ECFP4 = pd.read_csv('ECFP4_kampo_docking_result_1220.csv')
vina_affinity_kampo_ECFP4 = df_kampo_ECFP4['vina_affinity']
pic50_kampo_ECFP4 = df_kampo_ECFP4['pIC50']
p_ECFP4 = len(vina_affinity_kampo_ECFP4[vina_affinity_kampo_ECFP4 <= -8.0])/ len(vina_affinity_kampo_ECFP4)
print(f"total: {len(vina_affinity_kampo_ECFP4)}")
print(f"fit rate of ECFP4 model is {p_ECFP4}")

df_kampo_rdkit = pd.read_csv('RDKit_kampo_docking_result_1220.csv')
vina_affinity_kampo_rdkit = df_kampo_rdkit['vina_affinity']
pic50_kampo_rdkit = df_kampo_rdkit['pIC50']
p_rdkit = len(vina_affinity_kampo_rdkit[vina_affinity_kampo_rdkit <= -8.0])/ len(vina_affinity_kampo_rdkit)
print(f"total: {len(vina_affinity_kampo_rdkit)}")
print(f"fit rate of RDKit model is {p_rdkit}")
```

```python
p_ESI = len(vina_affinity_kampo_ESI[vina_affinity_kampo_ESI <= -8.0])/ len(vina_affinity_kampo_ESI)
print(f"fit rate of ESI model is {p_ESI}")
```

```python
# 説明変数と目的変数のデータ点の散布図をプロット
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(1, 1, 1)
fig.suptitle('correlation', fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=15)

plt.scatter(vina_affinity_kampo_ESI, pic50_kampo_ESI, color='red',edgecolors="m", marker='*', label='ESI', s=450, alpha=0.7)
plt.scatter(vina_affinity_kampo_ECFP4, pic50_kampo_ECFP4, c='Orange', edgecolors='#a65628', marker='p', label='ECFP4', s=300, alpha=0.7)
plt.scatter(vina_affinity_kampo_rdkit, pic50_kampo_rdkit, c='c', edgecolors='Blue', marker='o', label='RDKIt', s=250,  alpha=0.7)
plt.scatter(vina_affinity_a_glu , pic50_a_glu, c='y', edgecolors='orange' , marker='o' ,label='Reported', s=100, alpha=0.7)

plt.xlabel('vina affinity(kcal/mol)', fontsize=20)  # x軸のラベル
plt.ylabel('pIC50(Pred)', fontsize=20)  # y軸のラベル
#plt.xlim(2.5, 7.5)
#plt.ylim(2.5, 7.5)
plt.legend(loc=1,fontsize=15)

plt.savefig('correlation_predpIC50_affinity_0111.png')
```
