```python
import os
import rdkit
from rdkit import Chem
import pandas as pd
from rdkit.Chem import AllChem, Draw, PandasTools
import vina
import oddt
from oddt import toolkit
from oddt.docking import AutodockVina
```

```python
df = pd.read_csv('kampo_virtual_screening_result_1013_checked.csv')
df = df.sort_values('pIC50', ascending=False)
top = df[:5]

smiles_list = top['smiles'].to_list()
mols_list = [Chem.MolFromSmiles(smile) for smile in smiles_list]

numbers = top['number'].to_list()
```

```python
lig_list = numbers
len(lig_list)
```

```python
protein = next(oddt.toolkit.readfile('pdb', '3a4a.pdb'))
protein.protein = True
type(protein)
```

```python
for lig in lig_list:
    print(lig)
    ligand = next(oddt.toolkit.readfile('sdf', 'kampo_ligand_sdf/' + lig + '.sdf'))
    
    docking = AutodockVina.autodock_vina(protein=protein, size=(70, 80 ,80), center=(25, -2, 17), exhaustiveness=32, num_modes=9, energy_range=3, n_cpu=32)
    dock_result = docking.dock(ligand)
    
    SDFile = oddt.toolkit.Outputfile('sdf', 'kampo_output/3a4a_' + lig + '_out.sdf', overwrite=True)
    for mol in dock_result:
        SDFile.write(mol)
    SDFile.close()
    print(f"Docking {lig} to 3A4A is Done")
```
