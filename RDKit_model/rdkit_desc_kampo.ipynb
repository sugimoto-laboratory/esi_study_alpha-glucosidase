```python
import numpy as np
import pandas as pd
import rdkit 
from matplotlib import cm
from matplotlib import pyplot as plt
from rdkit.Chem import Descriptors
```

```python
desc_list =Descriptors.descList
print(len(desc_list))
```

```python
df = pd.read_csv('../../kampo_db/kampo_data_original/230922/check/kampo_ESI_tateishi_1004_checked.csv')
train = pd.read_csv('231202_tateishi_a-glucosidase_rdkit.csv')
all_variables =train.columns.values

smiles = df['smiles'].to_list()
mols = []
for i in smiles:
    mol = rdkit.Chem.MolFromSmiles(i)
    mols.append(mol)

df_mol = pd.DataFrame(mols)
df_mol.columns = ["ROMol"]
df_mol
```

```python
all_variables =train.columns.values[6:]
len(all_variables)
```

```python
all_variables
```

```python
rdkit_desc = pd.DataFrame()
for i, j in Descriptors.descList:
    rdkit_desc[i] = df_mol["ROMol"].map(j)
```

```python
rdkit_desc
```

```python
rdkit_desc_ =pd.concat([df[["number", "cid", "name", "smiles"]], rdkit_desc[all_variables]], axis=1)
rdkit_desc_
```
